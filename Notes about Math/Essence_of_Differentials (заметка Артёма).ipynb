## Часть 1.

⠀⠀⠀⠀Пусть $f$ — это некая функция от $x$: $f=f(x)$

Тогда её дифференицал $df$ — это **функция от двух переменных** $df = df(x, dx)$, которая выражается через переменные $x$ (аргумент) и $dx$ (приращение аргумента) следующим образом:

$$ {\displaystyle df = φ(x)·dx} $$

*Значение* дифференциала в данной точке $x_0$ показывает величину линейного приращения функции $f(x)$ в этой точке при заданном приращении аргумента $dx$:



Как видно из рисунка, значение *дифференциала* в данной точке *зависит* от того, какой величины мы выберем приращение $dx$. Но так как эта зависимость линейная (выражается через умножение), то увеличение $dx$, скажем, в три раза приведёт и к увеличению $df$ в те же три раза. А из этого следует, что значение функции $ {\displaystyle φ(x) = \frac{df}{dx} = \frac{3·df}{3·dx} = \frac{α·df}{α·dx} = \dots} $, связывающей приращение функции с приращением аргумента, в данной точке постоянно, т.е. $φ(x)$ действительно зависит только от $x$ (и не от $dx$).

⠀

⠀⠀⠀⠀Фунцию $φ(x)$ мы называем *производной* от функции $f(x)$ и обозначаем $f'(x)$. В каждой точке она показывает скорость роста своей первообразной $f(x)$, или, что то же самое, отношение её дифференциала к приращению аргумента. И в этом смысле становится понятно, почему, записывая равенство

$$ {\displaystyle f'(x) = \frac{df}{dx}} $$

Мы не указываем стремления $dx$ к нулю: оно к нему стремиться и не обязано. В качестве приращения аргумента мы можем взять какое угодно значение, но благодаря свойству линейности отношение $ {\displaystyle \frac{df}{dx}} $ останется тем же. Это как тангенс в подобных треугольниках (тут будет не грех вспомнить о знаменитом равенстве между производной и тангенсом угла наклона касательной — собственно, это он и есть).

Однако при реальном вычислении производной брать $dx$ каким угодно — плохая идея: всё-таки мы работаем с функцией, а не с касательной к ней, и насколько в общем случае изменится $df$ при произвольном увеличении $dx$, мы заранее не знаем (об этом нам как раз и должна бы сообщить искомая производная). К счастью, проблему решает устремление $dx$ к нулю: в этом случае разница между *линейным* приращением функции $df$ и *реальным* приращением $Δf$ становится ничтожно малой (по крайней мере, бесконечно малой по сравнению с $dx$). так что «неизвестный» дифференциал, связанный с касательной, становится возможно сделать «известным» — связать с функцией, расписав как разность её значений:

$$ {\displaystyle f'(x)=\frac{df}{dx}={\mathop{\lim }_{dx\to 0} \frac{\Delta f}{dx}\ }={\mathop{\lim }_{dx\to 0} \frac{f(x+dx)-f(x)}{dx}\ }} $$

Вроде бы обычно поступают наоборот — берут произвольное приращение аргумента $Δx$ и показывают, что при устремлении его к нулю реальное приращение функции совпадает с линейным. Но фишка тут как раз в том, что рассуждения работают в обе стороны; а между $dx$ и $Δx$, по-моему, вообще нет никакой разницы)

И ещё пара важных замечаний. Первое: дифференциал не существует сам по себе. В каком бы матемтаическом выражении мы ни работали с $df$, всегда, явно или неявно, подразумевается существование в том же выражении зависимой переменной (той, от которой зависит сама функция $f$ — обычно это $x$) и приращения $dx$ по этой переменной. И второе: если имеется зависимость вида не $f = f(x)$, а $f = f(u)$, где $u = u(x)$ — тоже какая-то функция, то форма записи дифференциала $df$ не меняется:

$$ {\displaystyle df = \frac{df}{dx}dx = \frac{df}{du} \frac{du}{dx} dx = f'(u) \frac{du}{dx} dx = f'(u) du} $$

(это свойство называется инвариантностью формы дифференциала)

## Часть 2.

⠀⠀⠀⠀Из предыдущих рассуждений вытекает, что операция взятия дифференциала — это операция нахождения по данной функции её линейных частей приращения (выраженных через $f'(x)$ и произвольное $dx$) в каждой точке. Тогда операцию взятия интеграла определим как *обратную к взятию дифференциала* — как восстановление функции по её линейным частям приращения в каждой точке:

$$ {\displaystyle \int{df}=f(x)+C} $$

Вполне естественно, что по одной только информации о линейных приращениях мы можем восстановить исходную функцию лишь с точностью до плюс константы — ведь абсолютно никакой информации хотя бы об одном *значении* $f$ линейные приращения не содержат — только о законе *изменения* этих значений.

«Классическое» определение интеграла из моего вытекает тут же — достаточно вспомнить, что дифференциал функции расписывается, собственно, как $df = f'(x)dx:$

$$ {\displaystyle \int{df} = \int{f'(x)dx} = f(x) + C} $$

Или, что то же самое:

$$ {\displaystyle \int{dF} = \int{f(x)dx} = F(x) + C} $$

Только теперь становится предельно ясно, откуда берётся $dx$ в подинтегральном выражении — из определения дифференциала, — и говорит оно просто-напросто о том, что подинтегральное выражение как математический объект остаётся дифференциалом. Совершенно естественным образом из этого следует, что вот это самое $dx$ в подинтегральном выражении является полноправным сомножителем, а не просто плейсхолдером — оно точно так же являлось полноправным сомножителем в обычной записи дифференциала, отсюда и вся история с линейной зависимостью.

Из инвариантности записи дифференциала следует, что нам неважно, по какой переменной или по какому выражению раскрывать первоначальный дифференциал:

$$ {\displaystyle \int{dF} = \int{f(x)dx} = \int{f(u)du} = \int{f(x^2-5)d(x^2-5)} = \dots } $$

А отсюда возникает возможность совершения и обратных действий, т.е. вот эти все плюшки в методах решения интегралов, когда мы подогняем целое выражение под табличный интеграл и с чистой совестью его берём:

$$ {\displaystyle \int{ \frac{1}{2x-1}dx } = \int{ \frac{1}{2x-1} \frac{d(2x-1)}{(2x-1)'} } = \frac{1}{2} \int{ \frac{1}{2x-1} d(2x-1) } = \frac{1}{2} ({\rm ln}|2x-1| + C) = \frac{1}{2} {\rm ln}|2x-1| + C } $$

Разумеется, мы можем брать дифференциал и от интеграла — интеграл ведь есть семейство функций с одинаковым законом изменения значений:

$$ {\displaystyle d\int{f(x)dx} = d(F(x) + C) = (F(x) + C)'dx = f(x)dx} $$

И, судя по всему, дифференциал от дифференциала также не возбраняется, ведь дифференциал — тоже функция, а от функции можно взять дифференциал. Правда, функция-функция — это всё-таки немного другая песня, чем функция-дифференциал, в последней обычно вот это вот $dx$ присутствует..:

$$ {\displaystyle d(df) = d(f'(x)dx)} $$

Но если все производные брать как частные производные по иксу (которыми они и являются), соответственно рассматривая при этом $dx$ как константу, то всё тоже просто великолепно сходится:

$$ {\displaystyle d(df) = d(f'(x)dx) = [f'(x)dx]'dx = [(f'(x))'dx + f'(x)(dx)']dx = [f''(x)dx + 0]dx = f''(x)dxdx = f''(x)dx^2 } $$
