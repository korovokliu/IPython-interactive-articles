# QUIZ 1

```python
A = {1, 2, 6}
B = {'a', 'b', 'c', 'd'}
C = {0, False}
print('A×B×C = ', [(a, b, c) for a in A for b in B for c in C])
print('(A×B)×C = ', [((a, b), c) for a in A for b in B for c in C])
```

# QUIZ 2

```python
Вопросы:
1) Пусть
А = {2, 3, 4}
B = {6, 8, 10}
R ⊂ A×B = {(a, b)∈ R  → y/x ∈ ℤ | a ∈ A, b ∈ B}
Какое из перечисленных множеств равно подмножеству R

</span>
Отношение <b>R</b> - это "если разделить элемент из A на элемент из B, получится целое число". 

</div>
```

```python
# Выведем декартово произведение множеств с помощью Python
A = {2, 3, 4}
B = {6, 8, 10}
print({(a, b) for a in A for b in B})
```

<img style="display: float; float: left; padding-right: 3px; max-height: 120px; width: auto;" src="relation01.png">
Теперь проверяем каждую пару:<br>
4<strike>  <b>R</b>  </strike>10, т.к. 10/4 = 2.5 ∈ ℚ<br>
3<b> R </b>6, т.к. 6/3=2 ∈ ℤ<br>
2<b> R </b>8, т.к. 8/2=4 ∈ ℤ<br>
...<br>
Код, которые выводит все упорядоченные пары, где это свойство выполняется:

```python
print({(x, y) for x in A for y in B if y%x==0})
```

# QUIZ 4

Вопросы:
1) Является ли отношение ⊆ транзитивным?<br>
<img style="display: float; float: right; padding-left: 3px; width: 200px;" src="транзитивность.png">
Да, т.к. если A является подмножеством B, a B подмножеством C, то A так же является подмножеством C

2) Является ли отношение ∈ транзитивным?<br>
Нет. Пример: A = {0}, B = {0, 1, {0}}, С = {x, y, {0, 1, {0}}}. A ∈ B и B ∈ C, но A ∉ C (элементами С являются x, y и множество B, но не то, что внутри него. Элементы внутри B могут быть подмножеством C, но не элементами по определению.
