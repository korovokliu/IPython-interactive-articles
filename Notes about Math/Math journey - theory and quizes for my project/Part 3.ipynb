<ul>

    <li>$F$ — <b>поле</b>, элементы которого называются <b>скалярами</b> (это может поле рациональных чисел ℚ, вещественных ℝ или комплексных ℂ чисел)</li>
    <br>
    <li>Определена операция сложения векторов $V×V → V$, сопоставляющая каждой паре $(x, y) ∈ V$ единственный элемент множества $V$ </li>
    <br>
    <li>Определена операция умножения векторов на <u>скаляры</u> $F×V → V$, сопоставляющая каждому элементу $λ ∈ F$ и каждому элементу $x ∈ V$ единственный элемент множества $V$, обозначаемый $λ·x$ </li>
</ul>

# QUIZ 6

Если в линейном пространстве $V$ каждому вектору $v$ по некому правилу $А$ поставлен в соответсвие вектор $u$ того же пространства (т.е. это бинарное <i><b>отношение</b></i>, которое является <i><b>функцией</b></i>, у которой есть свойства <i><b>бинарной операции</b></i> [раз уж отображение идет на том же множестве $×: V×V → V$]), то говорят, что задана <span style="background: #80ff4f;"><b> векторная функция векторного аргумента $u = A(v)$ </b></span>

Данная функция называется <b>линейным преобразованием</b> (или <b>линейным оператором</b>), если для неё выполнены перечисленные свойства линейности (сложение друг с другом и умножение на скаляр):
$$A(v+w) = A(v) + A(w) \\
A(λv) = λA(v)$$

<u>Пример</u>:
Дано линейное пространство векторов-строк $\overline{v}(v₁, v₂) : v₁, v₂ ∈ ℝ$, в которых определены операции сложения векторов $\overline{v}+\overline{w} = (v₁, v₂) + (w₁, w₂) = (v₁+w₁, v₂+w₂)$ и умножения вектора на число $λ\overline{v} = λ(v₁, v₂) = (λv₁, λv₂)$. Доказать, что функция векторного аргумента является линейным преобразованием
<hr>
Проверям свойства линейности:<br>
<ol>
<li> ${\displaystyle A(\overline{v}+ \overline{w}) = 2(\overline{v} + \overline{w}) = 2\overline{v} + 2\overline{w}= A(\overline{v}) + A(\overline{w})}$<br>
<i>дистрибутивность умножения вектора на скаляр относительно сложения векторов - аксиома векторного пространства</i></li>
<br>
<li> ${\displaystyle A(λ\overline{v}) = 2(λ\overline{v}) = 2(λ)\overline{v} (λ·2)\overline{v} = λ·(2\overline{v}) = λA(\overline{v})}$<br>
    ассоциативность умножения на скаляр, коммутативность самих действительных чисел (аксиома поля), аксиома ассоциативности
    </li>
    </ol>

## Аксиомы скалярного произведения

Скалярное произведение - функция, которая паре векторов из пространства V ставит в соответсвие некое вещественное число
$<·; ·> : V×V→ℝ$. Эта функция должна удовлетворять следующим аксиомам:
1) Симметричность <x, y>·<y, x> ∀x, y · V
2) Линейность по первому аргументу <x+y, z> = <x, z> + <y, z> ∀x, y, z ∈ V (если на первом месте сумма двух векторов x+y, то можно расписать ее как сумму двух скалярных произведений) 
3) Так же линейность по первому аргументу дает возможность выносить с первого места коэффициент k ∈ ℝ <kx, y> k<x, y> ∀x, y ∈ V ∩ ∀k ∈ ℝ
4) Положительная определенность - если один и тот же аргумент на первом месте, то отношение получается скалярным квадратом <x; x> ≥ 0, скалярный квадрат не может быть отрицательным. Ноль при применении данной операции должен получаться только в том в случае, когда x - нулевой элемент <x, x> = 0 ⇔ x = 0

Пример:
1) Пространство геометрических векторов $<a, b> = |a||b|cos(\widehat{a, b})$

2) Стандартное скалярное произведение в арифметических пространствах
$$ ℝⁿ = \left\{ { \left(\begin{array}{c}x₁\\x₂\\ \vdots \\ xₙ\end{array}\right): xᵢ ∈ ℝ } \right\} <u, v> = \sum^{n}_{i=1}u_i, v_i$$
Т.е. координаты векторов покомпонентно перемножаются и все это складывается 

3) Пространство многочленов, степень которых не превышает n над полем вещественных чисел
