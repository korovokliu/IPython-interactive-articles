<b>Эллипс</b> (геометрическое определение) - <details style="cursor: pointer; border: 1px solid gray; padding: 1px; border-radius: 0px 40px 0px 0px; display: inline-block; padding: 2px 1px 1px 1px; box-shadow: -1px 1px 1px 1px black, 1px -1px 1px 1px black, -2px -1px 2px 2px #9A5DFF, 2px 1px 2px 2px #9A5DFF;"> <summary style="padding: 0px 7px 0px 0px;">геометрическое место точек</summary>множество точек плоскости</details> $M$, сумма расстояний от которых до заранее выбранных точек $A$ и $B$ постоянна. Такие точки $A$ и $B$ называются <b>фокусами</b>, при чем расстояние a между ними
$${\displaystyle |AM|+|BM|=2\cdot a}$$
При чем само расстояние от одного фокуса до другого меньше их суммы

$${\frac {x^{2}}{a^{2}}}+{\frac {y^{2}}{b^{2}}}=1$$
