## Конечный автомат
Часто используется в пограммировании AI
Finite = limited number of states
State = How something is in that moment (for ex. water has Ice, Gas, Liqid states)

Finite-state machine have **only one state active at a time**

How can we use it?
1. Draw all the states


2. Show transitions between states


Finite-state machine for a lightbulb:


```python
from random import randint
from time import clock

##============= state section ============================

State = type("State", (object), {}) # создает класс State без атрибутов, подробнее см. ниже

class LightOn(State):  #state1
    def Execute(self):
        print("Light is on!")
        
class LightOff(State):  #state2
    def Execute(self):
        print("Light is off!")
        
#============= transition section =========================

class Transition:
    def __init__(self, to_state):
        self.to_state = to_state
    def Execute(self):
        print("Transitioning...")
        
```
