В магазине электроники раз в месяц проводится распродажа. Из всех товаров выбирают K товаров с самой большой ценой и делают на них скидку в 20%. По заданной информации о цене каждого из товаров и количестве товаров, на которые будет скидка, определите цену самого дорогого товара, не участвующего в распродаже, а также целую часть от суммы всех скидок.<br><br>
<b>Входные и выходные данные.</b> В первой строке входного файла находятся два числа,записанные через пробел: N – общее количество цен (натуральное число, не превышающее 10 000) и K – количество товаров со скидкой. В следующих N строках находятся значения цены каждого из товаров (все числа натуральные, не превышающие 10 000), каждое в отдельной строке. Запишите в ответе два числа: сначала цену самого дорогого товара, не участвующего в распродаже, а затем целую часть от суммы всех скидок.
Пример входного файла:<br>
10 3<br>
1800<br>
3600<br>
3700<br>
800<br>
2600<br>
2500<br>
1800<br>
1500<br>
1900<br>
1200<br>
При таких исходных данных ответ должен содержать два числа – 2500 и 1980. Пояснение: скидка будет на товары стоимостью 3700, 3600, 2600. Тогда самый дорогой товар без скидки стоит 2500, а сумма скидок 740+720+520 = 1980. 

```python
f = open("text")
line = f.readline()
first = line.split() # делим первую строку на список элементов, разделенных пробелом 
# два элемента - количество цен, количество товаров со скидкой. По умолчанию это 
# пробел, но можно указать любой символ, по которому он бы делил, например, 
# line.split("-")
n = int(first[0])
k = int(first[1])
```

Тот же код можно сделать короче с помощью функции map()
Первым аргументом она принимает функцию, а вторым аргументом принимает итерируемый объект, к элементом которого применяет эту функцию. (Так же можно передать ей третий, четвертый и т.д. итерируемый объект, но чаще обходятся одним)

```python
def multiply(arg):
    arg = arg*10
    return arg

l = [2, 15, 32, 6]

l = map(multiply, l)
print(l)
for i in l:
    print(i)
```

```python
f = open("26-1_умск.txt")
line = f.readline()
n, k = map(int, line.split())
print(n, k)
price = []
for i in f:
    price.append(int(i))
price.sort(reverse=True) # сортирует по убыванию
first = price[k] # первое наибольшее число, на которое не будет скидки. 
# к-ый элемент, т.к. отсчет в списке начианается с нуля
second = 0
"""Осталось вывести сумму всех скидок (не цены товара после скидки, а значения 
самих скидок):"""
for i in range(k): # k не включительно, поэтому до k
    second += price[i]*0.2
print(first, int(second))
```

По итогам проведения олимпиады по программированию каждый участник получил
определённое количество баллов, различное для каждого ученика. По регламенту олимпиады победителя присуждают K лучших участников, а призёра присуждают M лучших участников, следующих за ними. По заданной информации о результатах каждого из участников определите минимальный балл призёра и минимальный балл победителя данной олимпиады.<br>
<b>Входные и выходные данные</b>: В первой строке входного файла находятся три числа, записанные через пробел: N – общее количество результатов учащихся (натуральное число, не превышающее 10 000), K – количество победителей, M – количество призёров. В следующих N строках находятся значения каждого из результатов (все числа натуральные, не превышающие 1000), каждое в отдельной строке. Запишите в ответе два числа: сначала минимальный балл призёра, а затем минимальный балл победителя данной олимпиады без пробела. Прикрепи не только краткий ответ, но и код, чтобы куратор мог его проверить.
Пример входного файла:<br>
10 2 4<br>
244<br>
39<br>
213<br>
108<br>
132<br>
18<br>
46<br>
52<br>
242<br>
179<br>
При таких исходных данных ответ должен содержать 2 числа – 108 и 242. Пояснение: победители набрали 244 и 242 балла, призёры набрали 213, 179, 132, 108 баллов. Тогда минимальный балл призёра 108, а минимальный балл победителя – 242

```python
f = open('26_5_умск.txt')
line = f.readline()
n, k, m = map(int, line.split())
scores = []
for i in f:
    scores.append(int(i))
scores.sort(reverse=True)
min_win_score = scores[k-1]
min_awardee_score = scores[k + m-1]
print(min_awardee_score, min_win_score, sep="")
```

```python
f = open("26-3 (1)_умск.txt")
line = f.readline()
n, k, m = map(int, line.split())
all_prices = []
for price in f:
    print(price)
    all_prices.append(int(price))
all_prices.sort(reverse=True)
low_premium = all_prices[m-1] # т.к. отсчет с нуля, то до m-1
budget_sum = 0
count = 0
for i in range(m, len(all_prices)):
    budget_sum += all_prices[i]
    count +=1
budget_sum = budget_sum/count
print(low_premium, int(budget_sum))
```

```python
def divis(n, m):
    return n%m==0

def expression(x, a):
    val = divis(x,a) or (not divis(x, 15)) or (not divis(x, 18))
    return val

for a in range(1, 1000):
    flag = True
    for x in range(1, 1000):
        if not expression(x, a):
            flag = False
            break
    if flag==True:
        print(a)
```
