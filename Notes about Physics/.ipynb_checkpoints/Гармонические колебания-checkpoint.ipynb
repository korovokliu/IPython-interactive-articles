Что такое колебания?
--------------------------------
Что вообще такое "колебание"? Колебаться может температура с течением суток, уровень моря в зависимости от положения Луны (приливы и отливы), спрос на пуховики в зависимости от сезона; колеблются качели относительно земли или напряженность электрического поля.

У всех этих явлений есть одно общее свойство: изменения повторяются в точности или приблизительно с течением времени. То есть у них есть какая-то **периодичность**.

**Колебание - изменение какой-либо величины в точности (или приблизительно) повторяющееся с течением времени.**

Можно измерять колебания с помощью различных величин:
- Колебание качелей: изменение **угла** отклонения от положения равновесия
- Колебание спроса: 
- Колебание напряженности поля:

В механических колебаниях величину, изменение которой рассматривают с течением времени - координата (для линейной - в метрах, для угловой - радианы или градусы)
